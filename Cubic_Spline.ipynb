```python
x,y=[],[]
import csv
with open('dat.csv','r') as f:
    reader = csv.reader(f)
    for row in reader:
        x.append(row[0])
        y.append(row[1])
```

```python
import matplotlib.pyplot as plt
```

```python
x=[float(i) for i in x]
```

```python
y=[float(i) for i in y]
```

```python
plt.plot(x,y)
```

```python
from scipy.interpolate import CubicSpline
```

```python
cs = CubicSpline(x, y)
```

```python
plt.plot(x,cs(x))
```

```python
#For interpolation between given data points.
import numpy as np
x_fine = np.arange(1,100.005,0.005)
```

```python
plt.plot(x_fine, cs(x_fine))
plt.xlim(0,100)
plt.ylim(0,1)
```
